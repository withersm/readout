```python
filenames = glob.glob('20240215_reprocessed_load_curve_data/[!bias][!.ipynb]*')

print(filenames)

bias_points = pd.DataFrame({'Bias Line':[],'LO':[],'Bias Point':[]})

for pickle_i in filenames:#['./20240215_reprocessed_load_curve_data/20240213_results_4250MHz_BL10_Cold.pkl']:#filenames:
    plt.figure()
    
    temp = pickle_i.split('_')[-1].split('.')[0]
    ASU_BL = int(pickle_i.split('_')[-2].split('L')[1])
    LO = pickle_i.split('_')[-3].split('M')[0]
    
    if temp == 'Cold':
        continue
    
    with open(pickle_i, 'rb') as f:
        data_i = pickle.load(f) 
    
    bias_array = np.array([])
    for ch in range(len(data_i['Vtes'])):
        if np.isnan(data_i['bps'][ch]).all():
            continue
        split_pt = int(len(data_i['bps'][ch])/1.5)
        #print(data_i['bps'][ch])
        #print(len(data_i['bps'][ch]))
        #print(split_pt)
        Psat_index = dm.find_nearest_idx(data_i['bps'][ch][:split_pt]*100, 50)
        #print(Psat_index)
        bias_pt = data_i['Ibias'][Psat_index]
        bias_array = np.append(bias_array,bias_pt)
        plt.plot(data_i['Ibias'],data_i['bps'][ch]*100)
        plt.plot(data_i['Ibias'][Psat_index],data_i['bps'][ch][Psat_index]*100,'*')
        plt.vlines(data_i['Ibias'][Psat_index],0,100)
        
    #print(bias_array)
    current_bias_point = pd.DataFrame({'Bias Line':[ASU_BL],'LO':[LO],'Bias Point':[np.nanmedian(bias_array)]})
    bias_points = bias_points.append(current_bias_point, ignore_index=True)
    
    #print(temp, ASU_BL)
    plt.ylim([-50,150])
    plt.title(pickle_i)
    
bias_points.sort_values('LO',inplace=True)
bias_points.to_csv('./20240215_reprocessed_load_curve_data/bias_point_summary.csv')
```
